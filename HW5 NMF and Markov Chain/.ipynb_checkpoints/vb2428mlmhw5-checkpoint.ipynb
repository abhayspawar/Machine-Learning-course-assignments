```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
teamscores = pd.read_csv("CFB2016_scores.csv",header=None)
```

```python
teamscores.columns = ['TeamA','PointsA','TeamB','PointsB']
```

```python
teamscores.head()
```

```python
teamnames = np.loadtxt("TeamNames.txt",delimiter='\t',dtype=str)
```

```python
M = np.zeros([760,760])
```

```python
for k in range(len(teamscores)):
    i = teamscores['TeamA'][k] - 1
    j = teamscores['TeamB'][k] - 1
    pi  = teamscores['PointsA'][k]
    pj = teamscores['PointsB'][k]
    ti = 0
    tj = 0
    if pi>pj:
        ti = 1
    else:
        tj = 1
    M[i][i] = M[i][i] + ti + (pi*1.0)/(pi+pj)
    M[i][j] = M[i][j] + tj + (pj*1.0)/(pi+pj)
    M[j][i] = M[j][i] + ti + (pi*1.0)/(pi+pj)
    M[j][j] = M[j][j] + tj + (pj*1.0)/(pi+pj)
```

```python
np.sum(M)
```

```python
Mcopy = M.copy()
from sklearn.preprocessing import normalize
for i in range(len(M)):
    Mcopy[i] = normalize(M[i][:,np.newaxis],axis=0,norm='l1').ravel()
```

```python
np.sum(Mcopy)
```

```python
M = Mcopy
```

```python
def fun(t):
    w = np.ones(760) * 1.0/760
    #w = np.random.uniform(0,1,760)
    for i in range(t):
        w = np.dot(w,M)
    wdf = pd.DataFrame({'weights':w})
    wdf['index'] = wdf.index
    wdf = wdf.sort_values('weights',ascending=False)
    return wdf[:25]
```

```python
t1 = fun(10)
t2 = fun(100)
t3 = fun(1000)
t4 = fun(10000)
```

```python
t1['names'] = teamnames[t1['index']]
t2['names'] = teamnames[t2['index']]
t3['names'] = teamnames[t3['index']]
t4['names'] = teamnames[t4['index']]
```

```python
t1.to_csv("t1.csv")
t2.to_csv("t2.csv")
t3.to_csv("t3.csv")
t4.to_csv("t4.csv")
```

```python
lambd,u = np.linalg.eig(M.T)
```

```python
lambd.argsort()
```

```python
winf = (u[:,3].T)/np.sum(u[:,3])
#winf3 = u[:,3].T/np.sum(u[:,3])
#winf8 = u[:,8].T/np.sum(u[:,8])
```

```python
norms = np.zeros(10000)
wtemp = np.ones(760) * 1.0/760
for t in range(10000):
    wtemp = np.dot(wtemp,M)
    norms[t] = np.linalg.norm(wtemp-winf,ord=1)
```

```python
plt.plot(range(10000),norms)
plt.xlabel("t")
plt.ylabel("||wt - winf||1")
```

```python
nyt = np.loadtxt("nyt_data.txt",dtype=str,delimiter='\t')
```

```python
X = np.zeros([3012,8447])
```

```python
for i in range(len(nyt)):
    for pair in nyt[i].split(','):
        X[int(pair.split(':')[0])-1][i] = int(pair.split(':')[1])
        
```

```python
W = np.zeros([3012,25])
H = np.zeros([25,8447])
```

```python
for i in range(3012):
    W[i] = np.random.uniform(1,2,25)
for i in range(8447):
    H[:,i] = np.random.uniform(1,2,25)
```

```python
obj = np.zeros(100)
for t in range(100):
    print t
    wh = np.dot(W,H)
    wh[wh==0] = 1e-16
    purple = X/wh
    pink = W.T.copy()
    for i in range(len(pink)):
        pink[i] = normalize(pink[i][:,np.newaxis],axis=0,norm='l1').ravel()
    H = H *np.dot(pink,purple)
    
    wh = np.dot(W,H)
    wh[wh==0] = 1e-16
    purple = X/wh
    
    lblue = H.T.copy()
    for j in range(lblue.shape[1]):
        lblue[:,j] = normalize(lblue[:,j][:,np.newaxis],axis=0,norm='l1').ravel()
    W = W * np.dot(purple,lblue)
    wh = np.dot(W,H)
    wh[wh==0] = 1e-16
    obj[t] = np.sum(X*-np.log(wh)+wh)
    print obj[t]
```

```python
plt.plot(range(100),obj)
plt.xlabel("t")
plt.ylabel("objective")
```
